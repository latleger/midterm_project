# Cleaned csv files for use in Tableau

```python
# data processing
import pandas as pd 
```

```python
# libraries
train_df = pd.read_csv('clean_train.csv')
test_df = pd.read_csv('clean_test.csv')
```

```python
train_df.head()
```

```python
def convert_ticket_class(data):
    """
    Convert the 'Ticket_class' column to string values.
    """
    ticket_class = { 1 : "1st class", 2 : "2nd class", 3 : "3rd class"}
    for dataset in data:
        dataset['Ticket_class'] = dataset['Ticket_class'].map(ticket_class)
        
def convert_sex(data):
    """
    Convert the 'Sex' column to string values.
    """
    genders = { 0 : "male", 1 : "female"}
    for dataset in data:
        dataset['Sex'] = dataset['Sex'].map(genders)

def convert_embarked(data):
    """
    Convert the 'Embarked' column to string values.
    """
    ports = {0 : "S", 1 : "C", 2 : "Q"}
    for dataset in data:
        dataset['Embarked'] = dataset['Embarked'].map(ports)
        
# Convert columns to string values
data = [train_df, test_df]
convert_ticket_class(data)
convert_sex(data)
convert_embarked(data)
```

```python
def convert_survival(data):
    """
    Convert the 'Survived' column to string values.
    """
    survival = { 0 : "no", 1 : "yes"}
    for dataset in data:
        dataset['Survived'] = dataset['Survived'].map(survival)
        
# Convert columns to string values
data = [train_df]
convert_survival(data)
```

```python
# imports
import pandas as pd
import folium
```

```python
# bike station info
bike_data = 'austin_stations.csv'
austin_stations = pd.read_csv(bike_data)

# points of interests
yelp_data = 'yelp_info.csv'
yelp_locations = pd.read_csv(yelp_data)

# transit stops
transit_data = 'Stops.csv'
transit_locations = pd.read_csv(transit_data)
fitered_locations = transit_locations[transit_locations['ZIP'] == 78701]
```

```python
# Create a map of downtown Austin
map = folium.Map(location=[30.2672, -97.7387], zoom_start=14)
```

```python
# Add markers for bike stations
for index, row in austin_stations.iterrows():
    latitude = row['Latitude']
    longitude = row['Longitude']
    coords = [latitude, longitude]
    
    # Create a custom icon with the desired color and "font awesome"
    icon = folium.Icon(color='red', icon='bicycle', prefix='fa')
    
    # Add the marker with the custom icon to the map
    folium.Marker(coords, popup=f"Bike Station\nAvailable Bikes: {row['Available_Bikes']}", icon=icon).add_to(map)
```

```python
# Add markers for points of interest
for index, row in yelp_locations.iterrows():
    latitude = row['yelp_latitude']
    longitude = row['yelp_longitude']
    coords = [latitude, longitude]
    folium.Marker(coords, popup=row['name']).add_to(map)
```

```python
# Add markers for bus stops
for index, row in fitered_locations.iterrows():
    latitude = row['LATITUDE']
    longitude = row['LONGITUDE']
    coords = [latitude, longitude]
    
    # Create a custom icon with the desired color and "font awesome"
    icon = folium.Icon(color='purple', icon='bus', prefix='fa')
    
    folium.Marker(coords, popup=f"Stop ID: {row['STOP_ID']}", icon=icon).add_to(map)
```

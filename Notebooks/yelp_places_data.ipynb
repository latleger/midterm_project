```python
import pandas as pd
import requests
import os

# Assuming you have a dataframe called 'austinbikes_df' with station data
yelp_key = os.environ['yelp_key2']

URL = "https://api.yelp.com/v3/businesses/search"

header_dict = {
    "Accept": "application/JSON",
    "Authorization": "Bearer " + yelp_key
}

a = []
b = []
c = []
d = []
e = []
f = []
g = []
h = []
j = []
m = []
n = []

    # Set parameters for Yelp API request
params = {
    "term": "restaurant, bars, entertainment",
    "location": "Austin, TX 78701",
    "categories": "bars, parking, entertainment"
}

response = requests.get(URL, params=params, headers=header_dict)
res_json = response.json()

try:
    for i in res_json["businesses"]:
        a.append(i.get("name", None))  
        if "location" in i and "address1" in i["location"]:
             b.append(i["location"]["address1"]) 
             j.append(i["location"]["zip_code"])
        else:
             b.append(None)
             j.append(None)
        c.append(i.get("distance", None))  
        if "categories" in i and isinstance(i["categories"], list) and len(i["categories"]) > 0:
            d.append(i["categories"][0].get("title", None))  
        else:
            d.append(None)
        e.append(i.get("review_count", None))  
        f.append(i.get("rating", None))  
        g.append(i["coordinates"].get("longitude", None)) 
        h.append(i["coordinates"].get("latitude", None)) 
        m.append(i.get("price", None))
        n.append(i.get("id"))
except:
    pass

yelp_df = pd.DataFrame({"name": a, "address": b, "distance": c, "category_name": d, "review_count": e, "rating": f, "longitude": g, "latitude": h, "postcode": j, "price": m, "id":n})

# Print or do further processing with the dataframe 'yelp_df'
print(yelp_df)
```
